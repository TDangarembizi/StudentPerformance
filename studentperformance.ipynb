```python
sns.countplot(y='Gender', data=data)
ax = sns.countplot(y='Gender', data=data)

ax.set_yticklabels(['Male', 'Female'])
```

```python
gender_counts = data["Gender"].value_counts()
labels = gender_counts.index.map({0: 'Male', 1: 'Female'})
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Gender')
plt.axis('equal')
plt.show()
```

```python
pEd_counts = data["ParentalEducation"].value_counts()
labels = pEd_counts.index.map({0: 'None',1: 'High School',2: 'Some College',3: 'Bachelors',4: 'Higher'})
plt.figure(figsize=(6, 6))
plt.pie(pEd_counts, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Parental Education')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['StudyTimeWeekly'], bins=10, edgecolor='black')
plt.title('Distribution of Weekly Hours Studied')
plt.xlabel('Hours Studied')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
gClass_counts = data["GradeClass"].value_counts()
labels = gClass_counts.index.map({0: "'A' (GPA >= 3.5)",1: "'B' (3.0 <= GPA < 3.5)",2: "'C' (2.5 <= GPA < 3.0)",3: "'D' (2.0 <= GPA < 2.5)",4: "'F' (GPA < 2.0)"})
plt.figure(figsize=(6, 6))
plt.pie(gClass_counts, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Grade Class')
plt.axis('equal')
plt.show()
```

```python
Y=class0[['GradeClass']]
Y.head()
```
